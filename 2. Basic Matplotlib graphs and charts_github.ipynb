# Basic Matplotlib graphs and charts

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#To avoid to include plt.show() everytime for plotting
%matplotlib inline
```

```python
x = np.arange(0,10)
y = np.arange(0,10)*x
print(x)
print(y)
```

# Line Graph

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'bo') # 'bo' - Blue Circled marker
```

```python
plt.plot(x,y,'go--') # 'bo' - Blue Circled marker
```

```python
# Alternatively, Above graph can be plotted with mentioning each parameter like below:
plt.plot(x,y,color='green', marker='o', linestyle='dashed')
```

```python
# We can additionally use few more parameters better visualization
plt.plot(x,y,color='green', marker='o', linestyle='dashed',linewidth=2, markersize=12)
```

# Sactter Plot

```python
plt.scatter(x,y,c='magenta',marker='*')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title('Example of Scatter Plot')
plt.yticks(y) # X Axis will show specific ticks as listed in y instead of default ticks
```

# Bar Plot

```python
x1 = [1,3,6,9] 
y1 = [1,9,36,81]  
x2 = [2,4,8,16] 
y2 = [4,8,16,32] 
plt.bar(x1, y1, color ='orange') 
plt.bar(x2, y2, color = 'black') 
plt.title('Example of Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis')
```

# Histogram

```python
data = np.arange(1,100,8) 
plt.hist(data,bins = 5) 
plt.title("Example of histogram") 
```

# Box Plot

```python
data = np.arange(1,100,8)
plt.boxplot(data)
```

# Pie Chart

```python
import matplotlib.pyplot as plt
player = ['Sachin','Ganguly','Dravid','Shewag']
run = ['90','80','101','67']
colors = ['orange','blue','silver','yellow']
plt.pie(x=run,labels=player,explode=[0,0,.25,0],colors=colors,autopct='%1.2f%%')
```

### Let's plot all these ghaph/charts in subplots using Axes object

```python
import numpy as np
x = np.arange(0,10)
y = np.arange(0,10)*x
print(x)
print(y)

fig = plt.figure(figsize=(20,12))
axes_line=fig.add_subplot(2,3,1)
axes_scatter=fig.add_subplot(2,3,2)
axes_bar=fig.add_subplot(2,3,3)
axes_hist=fig.add_subplot(2,3,4)
axes_box=fig.add_subplot(2,3,5)
axes_pie=fig.add_subplot(2,3,6)

axes_line.plot(x,y,'go')
axes_line.set_title('Line Graph')
axes_scatter.scatter(x,y,marker='*')
axes_scatter.set_title('Scatter Plot')
axes_bar.bar(x,y)
axes_bar.set_title('Bar Graph')
axes_hist.hist(y,bins=3)
axes_hist.set_title('Histogram')
axes_box.boxplot(x)
axes_box.set_title('Box Plot')
axes_pie.pie(y,autopct='%1.1f%%')
axes_pie.set_title('Pie Chart')
```

### Let's plot all these ghaph/charts in subplots using Axes object from subplot (without using Figure Class)

```python
import numpy as np
x = np.arange(0,10)
y = np.arange(0,10)*x
print(x)
print(y)

axes_line=plt.subplot(2,3,1)
axes_scatter=plt.subplot(2,3,2)
axes_bar=plt.subplot(2,3,3)
axes_hist=plt.subplot(2,3,4)
axes_box=plt.subplot(2,3,5)
axes_pie=plt.subplot(2,3,6)

axes_line.plot(x,y,'go')
axes_line.set_title('Line Graph')
axes_scatter.scatter(x,y,marker='*')
axes_scatter.set_title('Scatter Plot')
axes_bar.bar(x,y)
axes_bar.set_title('Bar Graph')
axes_hist.hist(y,bins=3)
axes_hist.set_title('Histogram')
axes_box.boxplot(x)
axes_box.set_title('Box Plot')
axes_pie.pie(y,autopct='%1.1f%%')
axes_pie.set_title('Pie Chart')
```
