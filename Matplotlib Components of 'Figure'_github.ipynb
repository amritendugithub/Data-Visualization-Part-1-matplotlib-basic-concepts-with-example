# Matplotlib - Components of 'Figure'

Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy

Matplotlib graphs your data on Figure(s).Each of which can contain one or more Axes (i.e., an area where points can be specified in terms of x-y coordinates. The most simple way of creating a figure with an axes is using 'pyplot.subplots()'. We can then use Axes.plot to draw some data on the axes

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()  # Create a figure containing a single axes.
ax.plot([1, 2, 3, 4], [1, 4, 2, 3])  # Plot some data on the axes.
```

In fact, you can do the same using matplotlib.pyplot module that performs that plot on the "current" axes, creating that axes (and its parent figure) if they don't exist yet. So the previous example can be written more shortly without using 'Axes'

```python
plt.plot([1, 2, 3, 4], [1, 4, 2, 3])  # Matplotlib plot
```

## Parts of a 'Figure'



The figure keeps track of all the child Axes, a smattering of 'special' artists (titles, figure legends, etc), and the canvas. A figure can contain any number of Axes, but will typically have at least one.

```python
#The easiest way to create a new figure is with pyplot:

fig = plt.figure()  # an empty figure with no Axes
fig, ax = plt.subplots()  # a figure with a single Axes
fig, axs = plt.subplots(2, 2)  # a figure with a 2x2 grid of Axes
```

It's convenient to create the axes together with the figure, but you can also add axes later on, allowing for more complex axes layouts.

### Figures, Subplots, Axes and Ticks

A "figure" in matplotlib means the whole window in the user interface. Within this figure there can be subplots. While "subplot" positions the plots in a regular grid, "axes" allows free placement within the figure. Both can be useful depending on your intention. We've already worked with figures and subplots without explicitly calling them. When we call plot, matplotlib calls gca() to get the current axes and gca in turn calls gcf() to get the current figure. If there is none it calls figure() to make one, strictly speaking, to make a subplot(111).

All plotting is with respect to Axes. An Axes is made up of Axis objects and many other things.An Axes must belong to a Figure.
Typically You'll set up a Figure , and then add an Axes to it using fig.add_axes() but in most case you will use subplot (subplot is just axes on a grid system)





#### Subplots
With subplot you can arrange plots in a regular grid. You need to specify the number of rows and columns and the number of the plot. 






#### Axes
Axes are very similar to subplots but allow placement of plots at any location in the figure. So if we want to put a smaller plot inside a bigger one we do so with axes.

##### Let's Explore with code

```python
# Creating 4 subplots in a grid using Pyplot.subplots() method
fig, ax = plt.subplots(2,2)
```

```python
# Alternative, we can use add_subplot() to create such subplots using Figure object
fig = plt.figure()
ax = fig.add_subplot(441) #
ax1 = fig.add_subplot(4,4,16) # add_subplot() takes parameter in comma separated format as well
ax.set_title('first subplot')
ax1.set_title('last subplot')
```

Above example shows how we can plot subplots based on grid location. We have added subplots of 4X4 matrix to Figure object where maximum of 16 subplots can be plotted; In this case we have plotted 2 subplots both are at extreme diagonal positions(1st and 16th location). Each of these subplots are type of Axes object and we can apply functions available in Axes object for each subplot

Similarly we can directly add axes to Figure object (you can plot axes any position inside the figure unlike subplot which can only plotted in grid location)

```python
fig= plt.figure()
axes1= fig.add_axes([0.1,0.1,0.8,0.8])
axes2= fig.add_axes([0.17,0.5,0.4,0.3])
```

fig.add_axes([0.1,0.1,0.8,0.8]), makes the figure 10% from the left of the figure, 10% from the bottom of the figure, 80% width of the figure, and 80% height of the figure.

From left to the right, the values of the add_axes() function is [left, bottom, width, height]

We then create a second axes, axes2.

This second axes is created with the line, axes2= fig.add_axes([0.17,0.5,0.4,0.3])

This second axes is created and placed 17% from the left of the figure, 50% from the bottom of the figure, takes up 40% of the width of the figure, and takes up 30% of the height of the figure. This places this axes within the larger axes.

##### If we need to plot single graph - we can directly use pyplot object without using Figure and Axes objects

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-np.pi,np.pi,256,endpoint=True)
x
```

```python
y1=np.sin(x)
y2=np.cos(x)
plt.plot(x,y1)
plt.plot(x,y2)
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel("x axis")
plt.xlabel("y axis")
plt.legend(['sin','cos'])
plt.xlim(-4,4)
plt.ylim(-2,2)
plt.xticks([-np.pi,0,np.pi])
plt.yticks([-1,0,1])
plt.title('Sin Cos Curve')
plt.grid()
```

#### Above can be achieved using Axes object - Lets do that

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y1)
ax.plot(x,y2)
ax.set_xlabel("x axis")
ax.set_xlabel("y axis")
ax.legend(['sin','cos'])
ax.set_xlim(-4,4)
ax.set_ylim(-2,2)
ax.set_xticks([-np.pi,0,np.pi])
ax.set_yticks([-1,0,1])
ax.set_title('Sin Cos Curve')
ax.grid()
```

####  Lets do the same in subplot using Axes object

```python
fig = plt.figure()
ax = fig.add_subplot(321) 
ax2 = fig.add_subplot(326) 
ax1 = fig.add_subplot(4,4,16) # add_subplot() takes parameter in comma separated format as well
ax.plot(x,y1)
ax.plot(x,y2)
ax.set_xlabel("x axis")
ax.set_xlabel("y axis")
ax.legend(['sin','cos'])
ax.set_xlim(-4,4)
ax.set_ylim(-2,2)
ax.set_xticks([-np.pi,0,np.pi])
ax.set_yticks([-1,0,1])
ax.set_title('Sin Cos Curve')
ax.grid()
```
